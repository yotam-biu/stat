```python
salary = np.array([4,5,6,3,6,4,7,8,9,7,5,2])
```

```python
print("mean:", salary.mean())
print("range:", salary.max() - salary.min())
print("var:", round(salary.var(),2))
print("std:", round(salary.std(),2))
```

```python
salary = np.array([4,5,6,3,6,4,7,8,9,7,5,2])*2
print("mean:", salary.mean())
print("range:", salary.max() - salary.min())
print("var:", round(salary.var(),2))
print("std:", round(salary.std(),2))
```
